```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import janitor
```

```python
data = pd.read_csv('../data/mushrooms.csv')
data = data.clean_names()
```

```python
# Population vs class 

cross_tab_prop = pd.crosstab(index=data['population'],
                             columns=data['class'],
                             normalize="index")

cross_tab_prop.plot(kind='bar', 
                    stacked=True, 
                    colormap='tab10', 
                    figsize=(10, 6))


for n, x in enumerate([*cross_tab_prop.index.values]):
    for (proportion, y_loc) in zip(cross_tab_prop.loc[x],
                                   cross_tab_prop.loc[x].cumsum()):
        if proportion != 0:
                
            plt.text(x=n - 0.17,
                    y=y_loc,
                    s=f'{np.round(proportion * 100, 1)}%', 
                    color="black",
                    fontsize=12,
                    fontweight="bold")

plt.xlabel("Population")
plt.ylabel("Proportion")
plt.xticks([0, 1, 2, 3, 4, 5],['abundant','clustered','numerous','scattered','several','solitary'], rotation = 0)
plt.legend(labels=['Edible', 'Poisonous'], loc=4, fontsize='small', fancybox=True)
plt.show() 
```

The main objective of our EDA is to find the variables that have the most effect on the mushroom class. Here we can see if the population is abundant or numerous then the class will definitely be edible.

```python
cross_tab_prop = pd.crosstab(index=data['stalk_root'],
                             columns=data['class'],
                             normalize="index")

cross_tab_prop.plot(kind='bar', 
                    stacked=True, 
                    colormap='tab10', 
                    figsize=(10, 6))


for n, x in enumerate([*cross_tab_prop.index.values]):
    for (proportion, y_loc) in zip(cross_tab_prop.loc[x],
                                   cross_tab_prop.loc[x].cumsum()):
        if proportion != 0:                
            plt.text(x=n - 0.17,
                    y=y_loc,
                    s=f'{np.round(proportion * 100, 1)}%', 
                    color="black",
                    fontsize=12,
                    fontweight="bold")

plt.xlabel("Stalk_root")
plt.ylabel("Proportion")
plt.xticks([0, 1, 2, 3, 4],['missing','bulbous','club', 'equal', 'rooted'],rotation = 0)
plt.legend(labels=['Edible', 'Poisonous'], loc=4, fontsize='small', fancybox=True)
plt.show() 
```

We see that if the mushroom has a club, equal, or rooted stalk root, then there is a low chance for it to be poisonous. However, if the root is bulbous then a chance of 49.2% is that the mushroom is poisonous. Moreover, in the dataset we have 2480 mushrooms with a missing stalk root.

However, __silky__ mushrooms have an average possibility of 93.9% for being poisonuous.
